```python
LambdaMass_Signal=tree["Lambda0_M"].array()
LambdaMass_Background=tree2["Lambda0_M"].array()

print(LambdaMass_Signal)
```

```python
plt.hist(LambdaMass_Signal, bins=120, range=[1020,1120])
plt.xlabel('Mass [MeV/c^2]')
plt.title('Signal Lambda Mass')
plt.show()
```

```python
plt.hist(LambdaMass_Signal, bins=120, range=[1020,1120], density=True)
plt.hist(LambdaMass_Background, bins=120, range=[1020,1120], density=True)
plt.xlabel('Mass [MeV/c^2]')
plt.show()
```
